# Teachers' Actions and Results

```python
import json
import numpy as np
from pathlib import Path
from sklearn.cluster import KMeans
import ipywidgets as widgets
from IPython.display import display
```

```python
DATA_DIR = Path("../data")

data = json.loads((DATA_DIR / "embeddings_hybrid.json").read_text())
split = json.loads((DATA_DIR / "split.json").read_text())
train_ids = set(split["train"])

action_facets, action_vectors = [], []
result_facets, result_vectors = [], []

for transcript_id, conv in data.items():
    if transcript_id not in train_ids:
        continue
    for moment in conv.get("key_moments", []):
        if moment.get("annotation_type") != "scaffolding":
            continue
        for text, vec in zip(moment.get("action_decomposed") or [], moment.get("action_embeddings") or []):
            action_facets.append(text)
            action_vectors.append(vec)
        for text, vec in zip(moment.get("result_decomposed") or [], moment.get("result_embeddings") or []):
            result_facets.append(text)
            result_vectors.append(vec)

action_vectors = np.array(action_vectors)
result_vectors = np.array(result_vectors)

print(f"Action facets: {len(action_facets):,}  shape: {action_vectors.shape}")
print(f"Result facets: {len(result_facets):,}  shape: {result_vectors.shape}")
```

```python
def cluster_facets(facets, vectors, k, name):
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels = km.fit_predict(vectors)
    clusters = {}
    for label, text in zip(labels, facets):
        clusters.setdefault(int(label), []).append(text)
    print(f"\n{'='*60}")
    print(f"{name.upper()} FACETS — k={k}  ({len(facets):,} total)")
    print(f"{'='*60}")
    for i in range(k):
        texts = clusters.get(i, [])
        print(f"\nCluster {i}  ({len(texts):,} facets)")
        for t in texts[:5]:
            print(f"  • {t}")
        if len(texts) > 5:
            print(f"  ... ({len(texts) - 5:,} more)")

k_slider = widgets.IntSlider(min=2, max=30, step=1, value=10, description="k",
                              continuous_update=False)

@widgets.interact(k=k_slider)
def run_clustering(k):
    cluster_facets(action_facets, action_vectors, k, "action")
    cluster_facets(result_facets, result_vectors, k, "result")
```
